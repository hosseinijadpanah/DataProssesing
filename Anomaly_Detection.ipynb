# Anomaly Detection
Anomaly Detection is a technique used in data analysis and machine learning to identify patterns or data points that deviate significantly from the norm or expected behavior within a dataset.

- Anomalies, also known as outliers, can represent errors, fraud, defects, or other unusual events in various domains.

Here are some commonly used algorithms for anomaly detection:

**Isolation Forest:**
It's efficient for high-dimensional data and can handle large datasets.

One-Class SVM (Support Vector Machine):
It learns the distribution of normal data points and detects anomalies as deviations from that distribution.

Local Outlier Factor (LOF):
 It computes the local density deviation of a data point with respect to its neighbors.

K-Nearest Neighbors (KNN):
 It measures the distance between data points and identifies those with the smallest number of neighbors.

Elliptic Envelope:
 It models the data as a Gaussian distribution and identifies anomalies based on Mahalanobis distances.

Autoencoders:
 A type of neural network that learns to encode data and detect anomalies based on reconstruction errors.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise):
It identifies dense regions of data points and flags points in sparse regions as anomalies.

Minimum Covariance Determinant (MCD):
 It estimates the covariance matrix of a dataset and identifies anomalies based on the Mahalanobis distance.

Principal Component Analysis (PCA):
 It reduces the dimensionality of data and detects anomalies based on the residual errors.

Robust Random Cut Forest:
 It's a scalable and parallelizable algorithm for anomaly detection in high-dimensional data.

Histogram-based Outlier Detection (HBOS):
 It uses histograms to model the distribution of normal data and identifies anomalies.

Statistical Process Control (SPC):
 Used in manufacturing to detect anomalies in production processes.

Credit Card Fraud Detection:
 Identifying fraudulent transactions in the financial industry.

Network Intrusion Detection:
 Detecting abnormal network behavior or cyberattacks.

Quality Control in Manufacturing:
 Identifying defective products on assembly lines.

Healthcare:
 Detecting unusual patient behavior or medical test results for disease diagnosis.

Energy Grid Monitoring:
 Identifying anomalies in power consumption or grid performance.

Environmental Monitoring:
Detecting pollution spikes or unusual climate patterns.

Retail:
 Identifying unusual purchasing patterns or fraudulent transactions.

Agriculture:
 Detecting crop diseases or anomalies in crop growth.

Oil and Gas:
 Identifying anomalies in drilling operations or pipeline monitoring.

Transportation:
 Detecting unusual traffic patterns or vehicle maintenance issues.

Security:
 Detecting unauthorized access or unusual behavior in computer systems



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Generate a fictitious water quality dataset for demonstration purposes.
# Replace this with your actual water quality data.
np.random.seed(42)
data = np.random.randn(300, 2)  # Assuming you have 2 features.

# Standardize the data.
scaler = StandardScaler()
data_std = scaler.fit_transform(data)

# Train an Isolation Forest model for anomaly detection.
clf = IsolationForest(contamination=0.1, random_state=42)
clf.fit(data_std)

# Predict anomalies (1 for inliers, -1 for outliers).
anomaly_scores = clf.predict(data_std)

# Identify anomalies in the dataset.
anomalies = data[anomaly_scores == -1]

# Create a scatter plot of the data points.
plt.scatter(data[:, 0], data[:, 1], c='blue', label='Normal Data', alpha=0.7)
plt.scatter(anomalies[:, 0], anomalies[:, 1], c='red', label='Anomalies', marker='x', s=100)
plt.legend()
plt.title('Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sine wave signal with added "bad pattern" noise.
np.random.seed(42)
t = np.linspace(0, 2 * np.pi, 1000)  # Time points
signal = np.sin(t) + 0.2 * np.random.randn(1000)  # Sine wave with noise

# Define a threshold to identify anomalies.
threshold = 1.5  # Adjust this threshold based on your data characteristics.

# Find anomalies (points that deviate significantly from the sine wave).
anomalies = np.where(np.abs(signal - np.sin(t)) > threshold)

# Create a plot to visualize the signal and anomalies.
plt.figure(figsize=(12, 6))
plt.plot(t, signal, label="Signal")
plt.scatter(t[anomalies], signal[anomalies], c='red', label="Anomalies")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Anomaly Detection in Signal")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sine wave signal with added "bad pattern" noise.
np.random.seed(42)
t = np.linspace(0, 2 * np.pi, 1000)  # Time points
signal = np.sin(t) + 0.6 * np.random.randn(1000)  # Sine wave with more pronounced noise

# Define a threshold to identify anomalies.
threshold = 0.5  # Adjust this threshold based on your data characteristics.

# Find anomalies (points that deviate significantly from the sine wave).
anomalies = np.where(np.abs(signal - np.sin(t)) > threshold)[0]

# Create a plot to visualize the signal and anomalies.
plt.figure(figsize=(12, 6))
plt.plot(t, signal, label="Signal")
plt.scatter(t[anomalies], signal[anomalies], c='red', label="Anomalies", s=50)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Anomaly Detection in Signal")
plt.legend()
plt.grid(True)
plt.show()
```
